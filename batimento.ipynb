```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
n = 50  # Número de ondas
t = np.linspace(0, 50 * np.pi, 1000000)  # Intervalo de tempo (temos que alterar o espaçamento e o tamanho do eixo)

# Inicializando a superposição
signal = np.zeros_like(t)
freq = 2 * np.pi  # Frequência base
df = 0.1  # Diferença mínima de frequência entre as ondas

# Somando n ondas com frequências muito próximas
i = 0
while i != n:
    frequency = freq + i * df  # Frequências ligeiramente diferentes
    signal += 1 * np.cos(frequency * t)
    i += 1

# Plotando o resultado
plt.figure(figsize=(10, 6))
plt.plot(t, signal)
plt.title(f'Superposição de {n} ondas com batimento')
plt.xlabel('t')
plt.ylabel('|E|')
plt.grid(True)
plt.savefig('bat-50.png', format='png')
plt.show()
```
