```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
n = 50  # Nmero de ondas
t = np.linspace(0, 50 * np.pi, 1000000)  # Intervalo de tempo (temos que alterar o espaamento e o tamanho do eixo)

# Inicializando a superposio
signal = np.zeros_like(t)
freq = 2 * np.pi  # Frequncia base
df = 0.1  # Diferena mnima de frequncia entre as ondas

# Somando n ondas com frequncias muito prximas
i = 0
while i != n:
    frequency = freq + i * df  # Frequncias ligeiramente diferentes
    signal += 1 * np.cos(frequency * t)
    i += 1

# Plotando o resultado
plt.figure(figsize=(10, 6))
plt.plot(t, signal)
plt.title(f'Superposio de {n} ondas com batimento')
plt.xlabel('t')
plt.ylabel('|E|')
plt.grid(True)
plt.savefig('bat-50.png', format='png')
plt.show()
```
